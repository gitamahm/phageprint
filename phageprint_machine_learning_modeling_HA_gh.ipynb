```python
#number of iterations per model
num_iters=10
#excluding one of the partners (04)
samples = ['06', '08', '12', '11', '01', '02', '07', '10', '09'] 

num_sample=len(samples)
colors = ['cornflowerblue', 'teal', 'rosybrown', 'maroon', 'darkslategray', 'coral', 'lavender', 'lightpink', 'purple', 'gold', '']
colors=colors[:num_sample]
cols=['LR', 'GBC', 'KNN', 'MLP']
iterations=list(range(num_iters))
time5 = shuffle(time4, random_state=1)

#excluding one of the partners
time6 = time5[time5['subject']!='04'] 
#ensuring that we take out the OTUs that are subject 4 specific and zero in other subjects
time6 = time6.loc[:,time6.sum(axis=0)!=0]

#additionally we want relative abundance of OTUs rather than their absolute counts
# Calculate the sum of counts per subject
subject_sums = time6.iloc[:,:-3].sum(axis=1)
# # Divide each count by the subject sum to calculate the relative abundance
rel = time6.iloc[:,:-3].div(subject_sums, axis=0)
rel['subject']=time6['subject']
#dataframes holding the aupr and auroc values 
pr_results = pd.DataFrame(columns=[x+'_'+str(it) for x in cols for it in iterations], index=[samples])
roc_results = pd.DataFrame(columns=[x+'_'+str(it) for x in cols for it in iterations], index=[samples])

fig, axs = plt.subplots(2, 4, figsize=(10,5))
plt.subplots_adjust(wspace=.5, hspace=.5)

#Looping through  iterations of model building (different splits of train and test)
for it in iterations:
    
    #looping through four models
    for m in [0,1,2,3]:
        if  m==0:
            model=LogisticRegression(solver='lbfgs', max_iter=300, random_state=it)
        elif m==1:
            model=GradientBoostingClassifier(random_state=it)
        elif m==2:
            model=KNeighborsClassifier()
        else:
            model=MLPClassifier( max_iter=300, solver='lbfgs')

        #looping through tissues
        for index,ct in enumerate(samples):
            
            #the sample of interest gets a positive label, and others get a zero
            rel.loc[rel['subject']==ct,'label']=1
            rel.loc[rel['subject']!=ct,'label']=0
#             df_sum = df3.sum(axis=1)
            #lets shuffle the dataframe so that the model isn't biased in anyway
            x= rel.iloc[:,:-2]
            y= rel.iloc[:,-1] #getting the subject labels for each sample
            x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=it)


            # fit a model
            model.fit(x_train, y_train)
            # predict probabilities
            yhat = model.predict_proba(x_test)
            # retrieve just the probabilities for the positive class
            pos_probs = yhat[:, 1]

            # calculate the no skill line as the proportion of the positive class
            no_skill = len(y[y==1]) / len(y)

            # calculate model precision-recall curve
            precision, recall, _ = precision_recall_curve(y_test, pos_probs)
            # plot the model precision-recall curve
            axs[0, m].plot(recall, precision,marker='.', color=colors[index], alpha=.4)
            axs[0, m].set_ylim(0, 1)
            # plot the no skill precision-recall curve
            axs[0, m].plot([0, 1], [no_skill, no_skill], linestyle='--', color='grey', alpha=.4)

            auc_score = auc(recall, precision)
            pr_results.loc[samples[index], cols[m]+'_'+str(it)]= np.round(auc_score,2)
            pr_results.loc['model', cols[m]+'_'+str(it)] = cols[m]
            pr_results.loc[samples[index], 'null'] = np.round(no_skill,2)

            # calculate roc curve for model
            fpr, tpr, _ = roc_curve(y_test, pos_probs)
            auc_score = auc(fpr, tpr)
            roc_results.loc[samples[index], cols[m]+'_'+str(it)]= np.round(auc_score,2)
            roc_results.loc['model', cols[m]+'_'+str(it)] = cols[m]

            # plot model roc curve
            axs[1,m].plot(fpr, tpr,marker='.', color=colors[index], label=samples[index], alpha=.4)

            # axis labels
            axs[0,0].set_ylabel('Precision')
            axs[0,m].set_xlabel('Recall')
            axs[1,0].set_ylabel('TPR')
            axs[1,m].set_xlabel('FPR')

            axs[0,0].set_title(cols[0])
            axs[0,1].set_title(cols[1])
            axs[0,2].set_title(cols[2])
            axs[0,3].set_title(cols[3])
            axs[1,m].plot([0, 1], [0, 1], linestyle='--', color='grey')    

axs[0,3].legend(bbox_to_anchor=(1, 1), frameon=False)
```

```python
# Define a function to compute the confidence interval
def compute_confidence_interval(x):
    if all(x == 1):
        return 1.0, 1.0
    else:
        confidence_interval = st.t.interval(0.95, len(x)-1, loc=x.mean(), scale=st.sem(x))
        return round(confidence_interval[0], 2), round(confidence_interval[1], 2)  # Round the confidence interval values to 2 decimal places

```

```python
rel2 = rel.iloc[:,:-2].T
sample_ids = rel2.columns
top= pd.DataFrame(index=sample_ids, columns=['top_otus','top_fraction']).reset_index()
top_otu_list = []
n=1 #top n OTUs in terms of their relative abundance in a sample
for s in sample_ids:
    #getting the top OTUs per sample
    otus=rel2.nlargest(n, s).index.tolist()
    #getting also a list of the top otus, not per sample, but in total
    top_otu_list.append(otus)
    top.loc[top['index']==s, 'top_otus']=[otus]
    #summing the total relative abundance taken by the top n OTUs to see how much of the data we we would be removing
    top.loc[top['index']==s,'top_fraction'] =rel2.loc[otus,s].sum()

#flattening out the list of lists, and taking the unique set
top_otu_list = list(set(list(itertools.chain.from_iterable(top_otu_list))))
print('number of top OTUs for all samples that will get excluded:', len(top_otu_list))
```

```python
#number of iterations per model
num_iters=10
#excluding one of the partners (04)
samples = ['06', '08', '12', '11', '01', '02', '07', '10', '09'] 

num_sample=len(samples)
colors = ['cornflowerblue', 'teal', 'rosybrown', 'maroon', 'darkslategray', 'coral', 'lavender', 'lightpink', 'purple', 'gold', '']
colors=colors[:num_sample]
cols=['LR']
iterations=list(range(num_iters))

#dataframes holding the aupr and auroc values 
pr_results = pd.DataFrame(columns=[x+'_'+str(it) for x in cols for it in iterations], index=[samples])
roc_results = pd.DataFrame(columns=[x+'_'+str(it) for x in cols for it in iterations], index=[samples])

fig, axs = plt.subplots(2, 1, figsize=(3,5))
m=0
model=LogisticRegression(solver='lbfgs', max_iter=300, random_state=it)


#Looping through  iterations of model building (different splits of train and test)
for it in iterations:
    
    #looping through samples
    for index,ct in enumerate(samples):

        #the sample of interest gets a positive label, and others get a zero
        rell8.loc[rell8['subject']==ct,'label']=1
        rell8.loc[rell8['subject']!=ct,'label']=0
#             df_sum = df3.sum(axis=1)
        #lets shuffle the dataframe so that the model isn't biased in anyway
        x= rell8.iloc[:,:-2]
        y= rell8.iloc[:,-1] #getting the subject labels for each sample
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=it)


        # fit a model
        model.fit(x_train, y_train)
        # predict probabilities
        yhat = model.predict_proba(x_test)
        # retrieve just the probabilities for the positive class
        pos_probs = yhat[:, 1]

        # calculate the no skill line as the proportion of the positive class
        no_skill = len(y[y==1]) / len(y)

        # calculate model precision-recall curve
        precision, recall, _ = precision_recall_curve(y_test, pos_probs)
        # plot the model precision-recall curve
        axs[0].plot(recall, precision,marker='.', color=colors[index], alpha=.4)
        axs[0].set_ylim(0, 1)
        # plot the no skill precision-recall curve
        axs[0].plot([0, 1], [no_skill, no_skill], linestyle='--', color='grey', alpha=.4)

        auc_score = auc(recall, precision)
        pr_results.loc[samples[index], cols[m]+'_'+str(it)]= np.round(auc_score,2)
        pr_results.loc['model', cols[m]+'_'+str(it)] = cols[m]
        pr_results.loc[samples[index], 'null'] = np.round(no_skill,2)

        # calculate roc curve for model
        fpr, tpr, _ = roc_curve(y_test, pos_probs)
        auc_score = auc(fpr, tpr)
        roc_results.loc[samples[index], cols[m]+'_'+str(it)]= np.round(auc_score,2)
        roc_results.loc['model', cols[m]+'_'+str(it)] = cols[m]

        # plot model roc curve
        axs[1].plot(fpr, tpr,marker='.', color=colors[index], label=samples[index], alpha=.4)

        # axis labels
        axs[0].set_ylabel('Precision')
        axs[0].set_xlabel('Recall')
        axs[1].set_ylabel('TPR')
        axs[1].set_xlabel('FPR')

        axs[0].set_title(cols[0])
#         axs[0,1].set_title(cols[1])
#         axs[0,2].set_title(cols[2])
#         axs[0,3].set_title(cols[3])
        axs[1].plot([0, 1], [0, 1], linestyle='--', color='grey')    
plt.tight_layout()

# axs[0,3].legend(bbox_to_anchor=(1, 1), frameon=False)
```
